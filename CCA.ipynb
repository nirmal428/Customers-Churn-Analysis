```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/content/Customer Churn.csv")
data.head()
```

```python
data.info()
```

```python
# replcing blank with 0 as tenure is 0 and no total charges are recorded
```

```python
data['TotalCharges']=data['TotalCharges'].replace(" ","0")
data['TotalCharges'] = data['TotalCharges'].astype(float)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
data["customerID"].duplicated().sum()
```

```python
def conv(value):
  if value == 1:
    return "Yes"
  else:
    return "No"
data["SeniorCitizen"]=data["SeniorCitizen"].apply(conv)
```

```python
data.head(10)
```

```python
ax = sns.countplot(x=data["Churn"],data=data)
ax.bar_label(ax.containers[0])
plt.title("Count of customers by Churn")
plt.show()
```

```python
plt.figure(figsize=(3,3))
gb=data.groupby("Churn").agg({"Churn":"count"})
plt.pie(gb["Churn"],labels=gb.index , autopct="%1.2f%%")
plt.title("% of churn costomer",fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(3,3))
sns.countplot(x="gender",data=data , hue="Churn")
plt.show()
```

```python
plt.figure(figsize=(3,3))
sns.countplot(x="SeniorCitizen",data=data , hue="Churn")
plt.show()
```

```python

plt.figure(figsize=(9,3))
sns.countplot(x="tenure",data=data,hue="Churn")
plt.show()
```

```python
plt.figure(figsize=(5,4))
ax = sns.countplot(x="Contract",data=data , hue="Churn")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
data.columns.values
```

```python


columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

fig, axes = plt.subplots(3, 3, figsize=(18, 15))
axes = axes.flatten()


for i, col in enumerate(columns):
    sns.countplot(data=data, x=col, ax=axes[i], palette='Set2',hue=data["Churn"])
    axes[i].set_title(col)
    axes[i].tick_params(axis='x', rotation=45)


plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(5,4))
ax = sns.countplot(x="PaymentMethod",data=data , hue="Churn")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.xticks(rotation=45)

plt.show()
```
